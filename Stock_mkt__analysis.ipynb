```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn
```

```python
stock = ["AAPL", "XPEV", "TSLA"]

stocks = yf.download(stock, start = "2020-01-01", end = "2024-06-08")
data = stocks.loc[:,"Close"].copy()
```

```python
data.plot(figsize =(17, 8), fontsize = 18)
plt.style.use("seaborn")
plt.title("Growth of the Three Stocks in the USA")
plt.show()
```
